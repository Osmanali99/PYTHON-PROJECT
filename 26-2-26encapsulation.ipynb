```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.__age = age

  def get_age(self):
    return self.__age

  def set_age(self, age):
    if age > 0:
      self.__age = age
    else:
      print("Age must be positive")

p1 = Person("Tobias", 25)
print(p1.get_age())

p1.set_age(26)
print(p1.get_age())
```

```python
class Student:
  def __init__(self, name):
    self.name = name
    self.__grade = 0

  def set_grade(self, grade):
    if 0 <= grade <= 100:
      self.__grade = grade
    else:
      print("Grade must be between 0 and 100")

  def get_grade(self):
    return self.__grade

  def get_status(self):
    if self.__grade >= 60:
      return "Passed"
    else:
      return "Failed"
    
    

student = Student("Emil")
student.set_grade(99)
print(student.get_grade())
print(student.get_status())

```

```python
class Calculator:
  def __init__(self):
    self.result = 0

  def __validate(self, num):
    if not isinstance(num, (int, float)):
      return False
    return True

  def add(self, num):
    if self.__validate(num):
      self.result += num
    else:
      print("Invalid number")

calc = Calculator()
calc.add(10)
calc.add(5)
calc.add(14)
calc.add(16) 
print(calc.result)
# calc.__validate(5) # This would cause an error
```

```python
#create a manger class who manage employee name salary team_size using seter and getter method:
class Manager:
    def __init__(self, name, salary, team_size):
        self.__name = name
        self.__salary = salary
        self.__team_size = team_size

    def set_name(self, name):
        self.__name = name

    def get_name(self):
        return self.__name

    def set_salary(self, salary):
        self.__salary = salary

    def get_salary(self):
        return self.__salary

    def set_team_size(self, team_size):
        self.__team_size = team_size

    def get_team_size(self):
        return self.__team_size
# Example usage
manager = Manager("Alice Johnson", 80000, 5)
print("Manager Name:", manager.get_name())
print("Manager Salary:", manager.get_salary())
print("Manager Team Size:", manager.get_team_size())

```
