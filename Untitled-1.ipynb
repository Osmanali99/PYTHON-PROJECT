```python
#create a overriding program triangle class:
class Triangle:
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height
# create a subclass of Triangle that overrides the area method to calculate the area of an equilateral triangle:
class EquilateralTriangle(Triangle):
    def __init__(self, side):
        super().__init__(side, (side * (3 ** 0.5)) / 2)

    def area(self):
        return (self.base ** 2 * (3 ** 0.5)) / 4
# create an instance of the EquilateralTriangle class and calculate its area:
equilateral_triangle = EquilateralTriangle(5)
print("Area of the equilateral triangle:", equilateral_triangle.area())

```

```python
#create a  triangle using method of overriding:
class Shape:
    def area(self):
        return 0

class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height

# Example usage:
triangle = Triangle(10, 5)
print(triangle.area())
```

```python

class Fibonacci:
    def __init__(self, n):
        self.n = n

    def sequence(self):
        fib_sequence = [0, 1]
        for i in range(2, self.n):
            next_fib = fib_sequence[i-1] + fib_sequence[i-2]
            fib_sequence.append(next_fib)
        return fib_sequence[:self.n]
# Example usage:
fib = Fibonacci(100)
print(fib.sequence())
```

```python
class fictorial:
    def __init__(self, n):
        self.n = n

    def calculate(self):
        if self.n == 0 or self.n == 1:
            return 1
        else:
            return self.n * factorial(self.n - 1)
# Example usage:
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
fictorial_instance = fictorial(5)
print(fictorial_instance.calculate())   
```

```python
#create a program to calculate the area of a circle using method of overriding:
import math
class Shape:
    def area(self):
        return 0
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * (self.radius ** 2)
# Example usage:
circle = Circle(5)
print("Area of the circle:", circle.area())
#using method of overriding to calculate the area of a rectangle:
class Shape:
    def area(self):
        return 0
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height
# Example usage:
rectangle = Rectangle(4, 5)
print("Area of the rectangle:", rectangle.area())
```

```python
#working of super() in python:
class Parent:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, I am {self.name}."  
class Child(Parent):
    def __init__(self, name, age):
        super().__init__(name)
        self.age = age

    def greet(self):
        parent_greeting = super().greet()
        return f"{parent_greeting} I am {self.age} years old."  
# Example usage:
child = Child("Alice", 10)
print(child.greet())
```

```python
#Create a base class Animal with method sound().
class Animal:
    def sound(self):
        list_of_sounds = ["Some sound", "Another sound", "Different sound"]
        print("Available sounds:", list_of_sounds)
# Create a subclass Dog that inherits from Animal and overrides the sound() method to print "Woof!".
class Dog(Animal):
    def sound(self):
        print("Woof!")
# Create a subclass Cat that inherits from Animal and overrides the sound() method to print "Meow!".
class Cat(Animal):
    def sound(self):
        print("Meow!")
# Create instances of Dog and Cat and call their sound() methods to demonstrate method overriding.
dog = Dog()
cat = Cat()
dog.sound()  # Output: Woof!
cat.sound()  # Output: Meow!

    
```
