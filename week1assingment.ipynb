```python
# Import the datetime module to get the current year
from datetime import date

# --- Step 1: Ask for the user's name ---
# input() displays a prompt and returns whatever the user types as a string.
name = input("Please enter your name: ")

# --- Step 2: Ask for the user's birth year ---
# We need an integer, so we wrap input() with int() to convert the string.
# Note: This will crash if the user enters something that isn't a whole number.
birth_year = int(input("Please enter your birth year: "))

# --- Step 3: Ask for the user's current age ---
# Again, convert the input string to an integer.
age = int(input("Please enter your current age: "))

# --- Step 4: Ask for a lucky number ---
# This time we want a floating‑point number (decimal allowed), so we use float().
lucky_number = float(input("Please enter your lucky number: "))

# --- Step 5: Calculate the year they will turn 100 ---
# Get the current year from the system.
current_year = date.today().year
# Use the hint formula: 100 - current age + current year.
year_turn_100 = 100 - age + current_year

# --- Step 6: Print a summary message ---
# We use f‑strings (formatted string literals) to embed variables directly.
# Numbers are automatically converted to strings inside the f‑string.
print("\n--- Summary ---")
print(f"Name: {name}")
print(f"Birth Year: {birth_year}")
print(f"Age: {age}")
print(f"Lucky Number: {lucky_number}")
print(f"You will turn 100 in the year: {year_turn_100}")
```

```python
# After the loop, ask the user for a number (assuming we have that)
user_number = int(input("Enter a number: "))

# Countdown using while loop
while user_number >= 1:
    if user_number > 10:   # If number > 10, stop countdown
        print("Number too large! Countdown aborted.")
        break
    print(user_number)
    user_number -= 1
else:
    # This else executes if loop finished without break (i.e., number <=10)
    print("Blast Off!")
```
