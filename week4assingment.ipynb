```python
# Perform merging, grouping, concatenation and binning operations in Pandas.

import pandas as pd
import numpy as np

employees = pd.DataFrame({
    'Employee_ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Name': ['Alice Johnson', 'Bob Smith', 'Charlie Brown', 'David Wilson', 
             'Eve Davis', 'Frank Miller', 'Grace Lee', 'Henry Taylor', 
             'Ivy Chen', 'Jack Jones'],
    'Age': [28, 35, 42, 31, 26, 38, 29, 45, 33, 27],
    'Department_ID': ['D01', 'D02', 'D01', 'D03', 'D02', 'D01', 'D03', 'D02', 'D01', 'D03'],
     'Salary': [85000, 72000, 95000, 68000, 78000, 88000, 71000, 92000, 82000, 65000]

})

# 2. DEPARTMENTS DATAFRAME
departments = pd.DataFrame({
    'Department_ID': ['D01', 'D02', 'D03', 'D04', 'D05'],
    'Department_Name': ['Engineering', 'Marketing', 'Sales', 'Human Resources', 'Finance'],
    'Location': ['New York', 'San Francisco', 'Chicago', 'Boston', 'Seattle'],
    'Manager_ID': [101, 105, 108, None, None]  # None for departments without manager
})

# 3. SALARIES DATAFRAME
salaries = pd.DataFrame({
    'Employee_ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Bonus': [5000, 3000, 7000, 2000, 4000, 6000, 2500, 8000, 4500, 1500],
    
})

print(employees)
print(departments)
print(salaries)
```

```python
pd.merge(employees, departments , how='inner')
```

```python
pd.merge(employees, departments, how='left')
```

```python
pd.merge(employees , departments , how='outer')
```

```python
new_employees = pd.DataFrame({
    'Employee_ID': [111, 112, 113, 114, 115],
    'Name': ['Kevin Martinez', 'Linda Chen', 'Michael Scott', 
             'Patricia Williams', 'Robert Johnson'],
    'Age': [29, 34, 41, 37, 32],
    'Department_ID': ['D04', 'D02', 'D01', 'D03', 'D05']
})


updated_employees = pd.concat([employees, new_employees], ignore_index=True)
print(updated_employees)
```

```python
pd.merge(employees , salaries)
```

```python
avg_salary = employees.groupby('Department_ID')['Salary'].mean()
avg_salary
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Sales.csv")
df
```

```python
sns.barplot(data=df , x="product", y="sales")
```
