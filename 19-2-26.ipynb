```python
def print_centered_triangle(rows):
    """
    Prints a centered triangle pattern using asterisks.

    Args:
        rows: The number of rows for the triangle (integer).
    """
    # The maximum width of the triangle (width of the last row) 
    # will be 2 * rows - 1 (or 2 * rows + 1 depending on implementation)
    # Using '2 * rows' as a sufficient centering width works well.
    width = 2 * rows 
    for i in range(1, rows + 1):
        # Calculate the number of stars for the current row (1, 3, 5, ...)
        stars = '*' * (2 * i - 1)
        # Use center() to center the stars within the total width
        print(stars.center(width))

# Example: Print a centered triangle with 5 rows
print_centered_triangle(5)
```
