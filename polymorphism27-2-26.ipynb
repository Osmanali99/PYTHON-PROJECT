```python
#create a base class of student and his derived class student roll number and student marks using polymorphism :
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def display_info(self):
        print(f"Student Name: {self.name}, Age: {self.age}")
class StudentRollNumber(Student):
    def __init__(self, name, age, roll_number):
        super().__init__(name, age)
        self.roll_number = roll_number
    def display_info(self):
        super().display_info()
        print(f"Roll Number: {self.roll_number}")
class StudentMarks(Student):
    def __init__(self, name, age, marks):
        super().__init__(name, age)
        self.marks = marks
    def display_info(self):
        super().display_info()
        print(f"Marks: {self.marks}")
# Create instances of StudentRollNumber and StudentMarks
student1 = StudentRollNumber("Alice", 20, "R123")
student2 = StudentMarks("Bob", 22, 85)
# Display information for both students
student1.display_info()
print()  # Add a newline for better readability
student2.display_info()
```

```python
class Car:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("Drive!")

class Boat:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("Sail!")

class Plane:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("Fly!")

car1 = Car("Ford", "Mustang")       #Create a Car object
boat1 = Boat("Ibiza", "Touring 20") #Create a Boat object
plane1 = Plane("Boeing", "747")     #Create a Plane object

for x in (car1, boat1, plane1):
  x.move()
```

```python
#create animal speak class with all concepts of inheritance and polymorphism encapsulation and abstraction :
#using abstract class and method
from abc import ABC, abstractmethod
class Animal(ABC):
    @abstractmethod
    def speak(self):
        pass
class Dog(Animal):
    def speak(self):
        return "Woof!"
class Cat(Animal):
    def speak(self):
        return "Meow!"
# Create instances of Dog and Cat
dog = Dog()
cat = Cat()
print(f"The dog says: {dog.speak()}")
print(f"The cat says: {cat.speak()}")
# using abstract class and method with encapsulation
from abc import ABC, abstractmethod
class Vehicle(ABC):
    def __init__(self, brand, model):
        self._brand = brand  # Encapsulated attribute
        self._model = model  # Encapsulated attribute

    @abstractmethod
    def move(self):
        pass

    def display_info(self):
        print(f"Brand: {self._brand}, Model: {self._model}")
class Car(Vehicle):
    def move(self):
        return "Drive!"
class Boat(Vehicle):
    def move(self):
        return "Sail!"
class Plane(Vehicle):
    def move(self):
        return "Fly!"
# Create instances of Car, Boat, and Plane
car1 = Car("Ford", "Mustang")
boat1 = Boat("Ibiza", "Touring 20")
plane1 = Plane("Boeing", "747")
# Display information and movement for each vehicle
for vehicle in (car1, boat1, plane1):
    vehicle.display_info()
    print(f"Movement: {vehicle.move()}")
# Create a base class of shape and its derived class circle and rectangle using polymorphism :
import math
class Shape:
    def __init__(self, color):
        self.color = color
    def area(self):
        pass  # Abstract method to be implemented by derived classes
class Circle(Shape):
    def __init__(self, color, radius):
        super().__init__(color)
        self.radius = radius
    def area(self):
        return math.pi * self.radius ** 2
class Rectangle(Shape):
    def __init__(self, color, width, height):
        super().__init__(color)
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
    
```
