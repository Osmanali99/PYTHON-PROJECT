```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.stats import norm
def plot_normal_distribution(mu, sigma):
    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
    y = norm.pdf(x, loc=mu, scale=sigma)
    plt.plot(x, y)
    plt.xlabel('Value')
    plt.ylabel('Probability Density')
    plt.title(f'Normal Distribution (μ={mu}, σ={sigma})')
    plt.grid(True)
    plt.show()
```

```python
#create a program find mean median mode standard deviation and quartiles:
import numpy as np
from scipy import stats
def calculate_statistics(data):
    mean = np.mean(data)
    median = np.median(data)
    mode = stats.mode(data)[0][0]
    std_dev = np.std(data)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    
    return {
        'mean': mean,
        'median': median,
        'mode': mode,
        'std_dev': std_dev,
        'q1': q1,
        'q3': q3
    }
# Example usage:    
data = [5, 31, 43, 48, 50, 41, 7, 11, 15, 39, 80, 82, 32, 2, 8, 6, 25, 36, 27, 61, 31]
statistics = calculate_statistics(data)
print(statistics)
print(f"Mean: {statistics['mean']}")
print(f"Median: {statistics['median']}")
print(f"Mode: {statistics['mode']}")
print(f"Standard Deviation: {statistics['std_dev']}")
print(f"Q1: {statistics['q1']}")
print(f"Q3: {statistics['q3']}")    
```

```python
#create single array 2d array and 3d array using numpy:
import numpy as np

# Single dimensional array
single_array = np.array([1, 2, 3, 4, 5])
print("Single Array:", single_array)

# Two dimensional array
two_d_array = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:")
print(two_d_array)

# Three dimensional array
three_d_array = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print("3D Array:")
print(three_d_array)
```

```python
print("Mean of single array:", np.mean(single_array))
print("Mean of 2D array:", np.mean(two_d_array))
```

```python
#print array size 
print("Size of single array:", single_array.size)
print("Size of 2D array:", two_d_array.size)
print("Size of 3D array:", three_d_array.size)
```
