```python
# Python
import pandas as pd

# Sample DataFrame with duplicate rows
data = {
    'Name': ['Alice', 'Bob', 'Alice', 'David', 'Bob'],
    'Age': [25, 30, 25, 40, 30],
    'City': ['Lahore', 'Karachi', 'Lahore', 'Islamabad', 'Karachi']
}

df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Step 1: Find duplicate rows
# This returns a boolean Series indicating duplicates (True for duplicate, False for first occurrence)
duplicates_mask = df.duplicated()
print("Duplicate mask (True means duplicate row):")
print(duplicates_mask)

# Step 2: Display only duplicates
duplicate_rows = df[df.duplicated()]
print("Duplicate rows:")
print(duplicate_rows)

# Step 3: Remove duplicate rows, keeping the first occurrence
df_no_duplicates = df.drop_duplicates()
print("DataFrame after removing duplicates (keeping first occurrence):")
print(df_no_duplicates)

# Optional: Remove duplicates while keeping the last occurrence
df_no_duplicates_last = df.drop_duplicates(keep='last')
print("DataFrame after removing duplicates (keeping last occurrence):")
print(df_no_duplicates_last)

# Optional: Remove duplicates based on specific columns only
df_no_duplicates_subset = df.drop_duplicates(subset=['Name', 'Age'])
print("DataFrame after removing duplicates based on 'Name' and 'Age':")
print(df_no_duplicates_subset)
```

```python
# Python
import pandas as pd

# Sample DataFrame
data = {
    'Task': ['Task A', 'Task B', 'Task C', 'Task D'],
    'Duration': [60, 120, 450, 30]  # 450 is likely incorrect
}
df = pd.DataFrame(data)

# Function to detect durations above a reasonable threshold (e.g., 240 minutes)
def detect_invalid_duration(duration, max_minutes=240):
    return duration > max_minutes

# Detect invalid durations
df['Invalid'] = df['Duration'].apply(detect_invalid_duration)

# Fix invalid durations: for demonstration, we'll set them to the median of valid values
median_valid_duration = df.loc[~df['Invalid'], 'Duration'].median()
df.loc[df['Invalid'], 'Duration'] = median_valid_duration

# Drop the helper column if desired
df.drop(columns='Invalid', inplace=True)

# Display the updated DataFrame
print(df)
```

```python
# Sample dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 22, 28],
    'Pulse': [80, 72, 90, 75]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display the original dataset
print("Original Dataset:")
print(df)

# Sort the dataset by 'Pulse' in ascending order
df_sorted = df.sort_values(by='Pulse', ascending=True)

# Reset index if desired
df_sorted.reset_index(drop=True, inplace=True)

# Display the sorted dataset
print("Dataset Sorted by Pulse (Ascending):")
print(df_sorted)
```

```python

# Sample dataset
data = {
    'Duration': [30, 45, 30, 60, 45, 60, 30],
    'Pulse': [80, 90, 85, 70, 88, 75, 82],
    'Calories': [200, 300, 250, 400, 350, 450, 220]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Group by 'Duration' and calculate average 'Pulse'
average_pulse = df.groupby('Duration')['Pulse'].mean().reset_index()

# Rename columns for clarity (optional)
average_pulse.rename(columns={'Pulse': 'Average_Pulse'}, inplace=True)

# Display the result
print(average_pulse)
```

```python
# Sample workout data
workouts = [
    {"name": "Running", "calories_burned": 500},
    {"name": "Cycling", "calories_burned": 450},
    {"name": "Swimming", "calories_burned": 600},
    {"name": "Jump Rope", "calories_burned": 550},
    {"name": "HIIT", "calories_burned": 650},
    {"name": "Yoga", "calories_burned": 200},
    {"name": "Pilates", "calories_burned": 250},
    {"name": "Rowing", "calories_burned": 480}
]

# Sort workouts by calories_burned in descending order
sorted_workouts = sorted(workouts, key=lambda x: x["calories_burned"], reverse=True)

# Get top 5 workouts
top_5_workouts = sorted_workouts[:5]

# Display the results
print("Top 5 Workouts Based on Calories Burned:")
for i, workout in enumerate(top_5_workouts, start=1):
    print(f"{i}. {workout['name']} - {workout['calories_burned']} Calories")
    
# Example Output:
# Top 5 Workouts Based on Calories Burned:
# 1. HIIT - 650 Calories
# 2. Swimming - 600 Calories
# 3. Jump Rope - 550 Calories
# 4. Running - 500 Calories
# 5. Rowing - 480 Calories
```
