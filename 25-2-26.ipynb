```python
class Restaurant:
    def __init__(self, name, admin_password):
        self.name = name
        self.admin_password = admin_password
        self.menu = []

    # Admin login
    def login(self, password):
        return password == self.admin_password

    # Add menu item
    def add_item(self, item_name, price, password):
        if self.login(password):
            self.menu.append({"item": item_name, "price": price})
            print(f"{item_name} added to menu.")
        else:
            print("Access denied. Wrong password.")

    # Update menu item
    def update_item(self, old_item, new_item, new_price, password):
        if self.login(password):
            for item in self.menu:
                if item["item"].lower() == old_item.lower():
                    item["item"] = new_item
                    item["price"] = new_price
                    print(f"{old_item} updated to {new_item}.")
                    return
            print(f"{old_item} not found in menu.")
        else:
            print("Access denied. Wrong password.")

    # Delete menu item
    def delete_item(self, item_name, password):
        if self.login(password):
            for item in self.menu:
                if item["item"].lower() == item_name.lower():
                    self.menu.remove(item)
                    print(f"{item_name} removed from menu.")
                    return
            print(f"{item_name} not found in menu.")
        else:
            print("Access denied. Wrong password.")

    # Search menu item
    def search_item(self, item_name):
        for item in self.menu:
            if item_name.lower() in item["item"].lower():
                print(f"Found: {item['item']} - Rs.{item['price']}")
                return
        print(f"{item_name} not found in menu.")

    # Customer view
    def show_menu(self):
        print(f"\nWelcome to {self.name}!")
        print("------ Menu ------")
        if not self.menu:
            print("Menu is empty.")
        else:
            for item in self.menu:
                print(f"{item['item']} - Rs.{item['price']}")
        print("------------------")

    # Save menu to file
    def save_menu(self, filename="menu.txt"):
        with open(filename, "w") as f:
            for item in self.menu:
                f.write(f"{item['item']} - Rs.{item['price']}\n")
        print("Menu saved to file.")

    # Load menu from file
    def load_menu(self, filename="menu.txt"):
        try:
            with open(filename, "r") as f:
                self.menu = []
                for line in f:
                    item, price = line.strip().split(" - Rs.")
                    self.menu.append({"item": item, "price": int(price)})
            print("Menu loaded from file.")
        except FileNotFoundError:
            print("No saved menu found.")


# Interactive usage
shop = Restaurant("Zahid Tika Shop", "admin123")

while True:
    print("\n--- Zahid Tika Shop ---")
    print("1. Admin Login")
    print("2. Customer View")
    print("3. Exit")
    choice = input("Enter choice: ")

    if choice == "1":
        pwd = input("Enter admin password: ")
        if shop.login(pwd):
            print("Admin logged in.")
            print("a. Add Item")
            print("b. Update Item")
            print("c. Delete Item")
            print("d. Save Menu")
            admin_choice = input("Enter choice: ")

            if admin_choice == "a":
                name = input("Item name: ")
                price = int(input("Price: "))
                shop.add_item(name, price, pwd)
            elif admin_choice == "b":
                old = input("Old item name: ")
                new = input("New item name: ")
                price = int(input("New price: "))
                shop.update_item(old, new, price, pwd)
            elif admin_choice == "c":
                name = input("Item name to delete: ")
                shop.delete_item(name, pwd)
            elif admin_choice == "d":
                shop.save_menu()
        else:
            print("Wrong password.")

    elif choice == "2":
        shop.show_menu()
        search = input("Search item (or press Enter to skip): ")
        if search:
            shop.search_item(search)

    elif choice == "3":
        print("Goodbye!")
        break

```

```python
#create online wallet class in which user can create account and add money and transfer money to other users
class OnlineWallet:
    def __init__(self):
        self.accounts = {}
        self.next_account_number = 1

    def create_account(self, name, initial_balance=0):
        account_number = self.next_account_number
        self.accounts[account_number] = {
            "name": name,
            "balance": initial_balance
        }
        self.next_account_number += 1
        print(f"Account created for {name}. Account Number: {account_number}")

    def add_money(self, account_number, amount):
        if account_number in self.accounts:
            self.accounts[account_number]["balance"] += amount
            print(f"Rs.{amount} added to account {account_number}")
        else:
            print("Account not found.")

    def transfer_money(self, from_account, to_account, amount):
        if from_account in self.accounts and to_account in self.accounts:
            if self.accounts[from_account]["balance"] >= amount:
                self.accounts[from_account]["balance"] -= amount
                self.accounts[to_account]["balance"] += amount
                print(f"Rs.{amount} transferred from account {from_account} to account {to_account}")
            else:
                print("Insufficient balance.")
        else:
            print("One or both accounts not found.")

    def show_balance(self, account_number):
        if account_number in self.accounts:
            balance = self.accounts[account_number]["balance"]
            name = self.accounts[account_number]["name"]
            print(f"Account Number: {account_number}, Name: {name}, Balance: Rs.{balance}")
        else:
            print("Account not found.")
    def input_username(self):
        self.username = input("Enter username: ")
        print(f"Username '{self.username}' entered.")
    def input_password(self):
        self.password = input("enter password: ")
        print(f"password '{self.password}' enterd")
            
            
# Example usage
wallet = OnlineWallet()
wallet.create_account("Alice", 1000)
wallet.create_account("Bob", 500)   
wallet.add_money(1, 200)  # Add money to Alice's account
wallet.transfer_money(1, 2, 300)  # Transfer money from Alice to Bob
wallet.show_balance(1)  # Show Alice's balance
wallet.show_balance(2) 
  # Get user input for login

wallet.input_username()
wallet.input_password()

 # Show Bob's balance
```

```python
#create username and password in security and login page
class Security:
    def __init__(self):
        self.users = {}

    def create_user(self, username, password):
        if username in self.users:
            print("Username already exists.")
        else:
            self.users[username] = password
            print(f"User '{username}' created successfully.")

    def login(self, username, password):
        if username in self.users and self.users[username] == password:
            print(f"Login successful. Welcome, {username}!")
        else:
            print("Login failed. Invalid username or password.")
# Example usage
security = Security()
security.create_user("admin", "admin123")
security.login("admin", "admin123")  # Successful login
```
