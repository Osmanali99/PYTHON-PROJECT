```python
#create a base class of student and his derived class student roll number and student marks using inheritance:
#base class student
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
#derived class student roll number
class StudentRollNumber(Student):
    def __init__(self, name, age, roll_number):
        super().__init__(name, age)
        self.roll_number = roll_number

    def display_roll_number(self):
        print(f"Roll Number: {self.roll_number}")
#derived class student marks
class StudentMarks(Student):
    def __init__(self, name, age, marks):
        super().__init__(name, age)
        self.marks = marks

    def display_marks(self):
        print(f"Marks: {self.marks}")
#creating an object of student roll number class
student1 = StudentRollNumber("Alice", 20, "R123")
student1.display_info()
student1.display_roll_number()
#creating an object of student marks class
student2 = StudentMarks("Bob", 22, 85)
student2.display_info()
student2.display_marks()
```

```python
#create a base class vehicle and derive a class car form it and class of motor cycle from it using inheritance:
#parent base class
class vehicle:
    def __init__(self, make, model):
        self.make = make
        self.model = model
    def display_info(self):
        print(f"Make: {self.make}, Model: {self.model}")
#derived class car
class car(vehicle):
    def __init__(self, make, model, num_doors):
        super().__init__(make, model)
        self.num_doors = num_doors
    def display_info(self):
        super().display_info()
        print(f"Number of doors: {self.num_doors}") 
#derived class motorcycle
class motorcycle(vehicle):
    def __init__(self, make, model, has_sidecar):
        super().__init__(make, model)
        self.has_sidecar = has_sidecar
    def display_info(self):
        super().display_info()
        print(f"Has sidecar: {self.has_sidecar}")
# Create instances of car and motorcycle
my_car = car("Toyota", "Camry", 4)
my_motorcycle = motorcycle("Harley-Davidson", "Street 750", False)
# Display information about the car and motorcycle
my_car.display_info()
my_motorcycle.display_info()
```

```python
# Parent/Base Class
class Vehicle:
    def __init__(self, make, model, year):
        """Initialize common vehicle attributes."""
        self.make = make
        self.model = model
        self.year = year
        self.started = False # Default attribute

    def start_engine(self):
        """Method to start the vehicle engine."""
        if not self.started:
            print(f"The {self.make} {self.model}'s engine started.")
            self.started = True
        else:
            print("The engine is already running.")

    def stop_engine(self):
        """Method to stop the vehicle engine."""
        if self.started:
            print(f"The {self.make} {self.model}'s engine stopped.")
            self.started = False
        else:
            print("The engine is already off.")

# Child/Derived Class
class Car(Vehicle):
    def __init__(self, make, model, year, num_doors):
        """Initialize car-specific attributes, while reusing parent's init."""
        # Call the parent class's __init__ method using super()
        super().__init__(make, model, year)
        self.num_doors = num_doors # New attribute specific to Car

    def drive(self):
        """Method specific to a car."""
        if self.started:
            print(f"The {self.model} is being driven with {self.num_doors} doors.")
        else:
            print(f"Can't drive the {self.model} yet, the engine is off.")
    
    # Optional: Override a parent method
    def start_engine(self):
        """Override the start_engine method to add car-specific behavior."""
        print("Checking car systems...")
        super().start_engine() # Call the parent's method


# Create an instance of the Car subclass
my_car = Car('Toyota', 'Camry', 2022, 4)

# Access attributes and methods from both the parent and child classes
print(f"My car is a {my_car.year} {my_car.make} {my_car.model} with {my_car.num_doors} doors.")

my_car.start_engine() # Calls the overridden method in Car, which then calls the Vehicle method
my_car.drive() # Calls the method specific to the Car class
my_car.stop_engine() # Calls the method inherited from the Vehicle class
```

```python
#create a shape class and derive a class of circle and rectangle from it using inheritance:
# Parent/Base Class
class Shape:
    def __init__(self, color):
        self.color = color
    def area(self):
        pass # This will be implemented in the derived classes
# Derived Class for Circle
class Circle(Shape):
    def __init__(self, color, radius):
        super().__init__(color) # Call the parent class's __init__ method
        self.radius = radius
    def area(self):
        import math
        return math.pi * (self.radius ** 2)
# Derived Class for Rectangle
class Rectangle(Shape):
    def __init__(self, color, width, height):
        super().__init__(color) # Call the parent class's __init__ method
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height


# Create instances of Circle and Rectangle
my_circle = Circle('red', 5)
my_rectangle = Rectangle('blue', 4, 6)
my_square = Rectangle('green', 3, 3) # A square is a rectangle with equal width and height

# Display the area of the circle, rectangle, and square
print(f"The area of the {my_square.color} square with side length {my_square.width} is: {my_square.area():.2f}")
print(f"The area of the {my_circle.color} circle with radius {my_circle.radius} is: {my_circle.area():.2f}")
print(f"The area of the {my_rectangle.color} rectangle with width {my_rectangle.width} and height {my_rectangle.height} is: {my_rectangle.area():.2f}")
```
